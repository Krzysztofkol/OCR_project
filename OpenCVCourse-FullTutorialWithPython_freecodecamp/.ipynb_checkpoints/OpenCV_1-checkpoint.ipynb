```python
import cv2 as cv
```

imread method - take the path of the image to return the matrix of pixelsd

```python
img = cv.imread('Resources/Photos/cat.jpg')
```

imshow method - displaying the image

```python
cv.imshow('Cat Picture',img)#name + the matrix
cv.waitKey(0)
```

```python
img = cv.imread('Resources/Photos/cat_large.jpg')
cv.imshow('Cat Large',img)#Size of the image > Size of the monitor
cv.waitKey(0)
```

Reading videos

```python
#capture =cv.VideoCapture(0)#0 - webcam; #1 - the second camera, ...
capture =cv.VideoCapture('Resources/Videos/dog.mp4')
#reading the video frame by frame
while True:
    isTrue, frame = capture.read()
    cv.imshow('Video',frame)
    #prevent the inf loop:
    if cv.waitKey(20)& 0xFF==ord('d'):
        break
capture.release()
cv.destroyAllWindows
cv.waitKey(0)
#(-215:Assertion failed) - because it run out of frames
```
