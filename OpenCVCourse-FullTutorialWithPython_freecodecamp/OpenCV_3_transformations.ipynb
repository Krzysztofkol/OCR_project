Translation - shifting an image across x & y axis

```python
#translation function
def translate(img,x,y):
    #translation matrix
    transMat=np.float32([[1,0,x],[0,1,y]])#2 lists inside
    #img.shape[1] - width
    #img.shape[0] - height
    dimensions=(img.shape[1],img.shape[0])
    return cv.warpAffine(img,transMat,dimensions)
#-x ->left
#-y ->up
#x ->right
#y->down
```

```python
translated=translate(img,250,250)
cv.imshow("Translated",translated)
cv.waitKey(0)
```

```python
untranslated=translate(translated,-250,-250)
cv.imshow("UnTranslated",untranslated)
cv.waitKey(0)
```

Rotation

```python
#rotating function: counterclockwise
def rotate(img, angle, rotPoint=None):
    (height,width)=img.shape[:2]#first(and only) to values!!! useful
    if rotPoint is None:
        rotPoint = (width//2,height//2)
    #rotation matrix
    rotMat=cv.getRotationMatrix2D(rotPoint,angle,1)#1 ~ scale
    dimensions=(width,height)
    return cv.warpAffine(img,rotMat,dimensions)
```

```python
rotated=rotate(img,45)
cv.imshow('Rotate',rotated)
cv.waitKey(0)
```

```python
rotated=rotate(img,-45)
cv.imshow('Rotate',rotated)
cv.waitKey(0)
```

```python
rotated_rotated=rotate(rotated,-45)
cv.imshow('Rotate Rotate',rotated_rotated)
cv.imshow('Rotate',rotated)
cv.waitKey(0)
```

Resizing the image

```python
resized = cv.resize(img, (500,500), interpolation=cv.INTER_CUBIC)
#INTER_AREA ~ shrinking
#INTER_LINEAR / INTER_CUBIC ~ enlarging the image
cv.imshow("Resized",resized)
cv.imshow("Image",img)
cv.waitKey(0)
```

```python
flipped = cv.flip(img,-1)#flipCode: 0- vertically, 1 - horizontally, -1 - both
cv.imshow("Flipped",flipped)
cv.imshow("Image",img)
cv.waitKey(0)
```
