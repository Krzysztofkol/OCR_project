Resize & rescale to ease off the computation strain

```python
#function
def rescaleFrame(frame,scale=0.2):
    #images, videos, live video
    width=int(frame.shape[1]*scale)
    height=int(frame.shape[0]*scale)
    dimensions=(width,height)
    return cv.resize(frame,dimensions,interpolation=cv.INTER_AREA)
```

```python
capture =cv.VideoCapture('Resources/Videos/dog.mp4')
while True:
    isTrue, frame = capture.read()
    
    frame_resized = rescaleFrame(frame)
    cv.imshow('Video',frame)
    cv.imshow('Video Resized',frame_resized)
    #prevent the inf loop:
    if cv.waitKey(20)& 0xFF==ord('d'):
        break
capture.release()
cv.destroyAllWindows()
```

```python
img = cv.imread('Resources/Photos/cat.jpg')
rezised_img=rescaleFrame(img)
cv.imshow('Cat Picture',rezised_img)#name + the matrix
cv.waitKey(0)
```

Paint a portion of an image

```python
blank=np.zeros((500,500,3),dtype='uint8')#3 ~ the number of colour channels!
blank[200:300,300:400]=0,255,0
blank[240:260,340:360]=0,0,255
cv.imshow('Image',blank)
cv.waitKey(0)
```

Draw a rectangle

```python
cv.rectangle(blank,(310,210),(390,290),(0,255,255),thickness=2)
cv.imshow('Rectanlge',blank)
cv.waitKey(0)
```

Draw a filled rectangle

```python
cv.rectangle(blank,(100,100),(400,400),(255,0,0),thickness=cv.Fi)
cv.imshow('Rectanlge',blank)
cv.waitKey(0)
```

A different way of specifying rectangle's coordinates

```python
blank=np.zeros((500,500,3),dtype='uint8')
cv.rectangle(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(0,255,0),thickness=1)
cv.imshow('Rectanlge',blank)
cv.waitKey(0)
```

```python
blank=np.zeros((500,500,3),dtype='uint8')
cv.rectangle(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(0,255,0),thickness=cv.FILLED)
cv.imshow('Rectanlge',blank)
cv.waitKey(0)
```

Draw a circle

```python
blank=np.zeros((500,500,3),dtype='uint8')
cv.circle(blank,(250,250),100,(0,255,0),thickness=cv.FILLED)
cv.imshow('Circle',blank)
cv.waitKey(0)
```

```python
blank=np.zeros((500,500,3),dtype='uint8')
cv.circle(blank,(blank.shape[1]//2,blank.shape[0]//2),100,(0,255,0),thickness=cv.FILLED)
cv.imshow('Circle',blank)
cv.waitKey(0)
```

Draw a line

```python
blank=np.zeros((500,500,3),dtype='uint8')
cv.line(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(255,255,255),thickness=3)
cv.imshow('Line',blank)
cv.waitKey(0)
```
