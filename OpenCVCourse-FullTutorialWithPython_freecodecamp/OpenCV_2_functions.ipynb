```python
import cv2 as cv
```

```python
img=cv.imread('Resources/Photos/cat.jpg')
cv.imshow('Cat',img)
cv.waitKey(0)
```

Convert to grayscale to see the distribution of colour instead of colour itself

```python
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('Gray',gray)
cv.waitKey(0)
```

```python
img=cv.imread('Resources/Photos/park.jpg')
cv.imshow('Park',img)
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('Gray',gray)
cv.waitKey(0)
```

Image blurring

Gaussian blurr

```python
blur=cv.GaussianBlur(img,(3,3),cv.BORDER_DEFAULT)#ksize - odd number!
cv.imshow('Park',img)
cv.imshow('Blur',blur)
cv.waitKey(0)
```

Edge Cascade - Finding edges present in the image

Canny

```python
canny=cv.Canny(img, 125,175)#2 treshold values
cv.imshow('Park',img)
cv.imshow('Canny',canny)
cv.waitKey(0)
```

Edge reduction by blurring the image
Slight blur ~ some of the edges

```python
canny=cv.Canny(blur, 125,175)#2 treshold values
cv.imshow('Park',img)
cv.imshow('Canny',canny)
cv.waitKey(0)
```

Dilating the image (thickening?)

```python
dilated=cv.dilate(canny,(3,3),iterations=3)#kernel; there can be several iterations at the time
cv.imshow('Dilate',dilated)
cv.imshow('Canny',canny)
cv.waitKey(0)
```

```python
Eroding the image (you can get the edge cascade)
```

```python
eroded=cv.erode(dilated,(3,3),iterations=3)
cv.imshow('Canny',canny)
cv.imshow('Dilate',dilated)
cv.imshow('Erode',eroded)
cv.waitKey(0)
```

Resize

```python
img=cv.imread('Resources/Photos/park.jpg')
resized=cv.resize(img,(500,500),interpolation=cv.INTER_CUBIC)#ignore tge aspect ratio
cv.imshow('Park',img)
#interpolation:
#shrinking the image to the smaller dimensions: interpolatiob=cv.INTER_AREA
#interpolatiob=cv.INTER_CUBIC~ slower, but higher quality
cv.imshow('Resized',resized)
cv.waitKey(0)
```

Cropping

Array Slicing

```python
cropped=img[50:200,200:400]
cv.imshow('Park',img)
cv.imshow("Cropped",cropped)
cv.waitKey(0)
```
